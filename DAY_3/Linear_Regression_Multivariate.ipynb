```python
import pandas as pd
import math
from sklearn import linear_model

# Load dataset
df = pd.read_csv("hiring.csv")

# Display dataset
print(df)

# Convert word numbers to integers
df.experience = df.experience.fillna("zero")

df.experience = df.experience.replace({
    'zero': 0,
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven': 7,
    'eight': 8,
    'nine': 9,
    'ten': 10,
    'eleven': 11
}).infer_objects(copy=False)

# Fill missing test score with median
median_test_score = math.floor(
    df['test_score(out of 10)'].median()
)

df['test_score(out of 10)'] = df[
    'test_score(out of 10)'
].fillna(median_test_score)

# Create Linear Regression model
reg = linear_model.LinearRegression()

# Train model
reg.fit(
    df[['experience',
        'test_score(out of 10)',
        'interview_score(out of 10)']],
    df['salary($)']
)

# Predict salary
prediction1 = reg.predict(
    pd.DataFrame(
        [[2, 9, 6]],
        columns=[
            'experience',
            'test_score(out of 10)',
            'interview_score(out of 10)'
        ]
    )
)

prediction2 = reg.predict(
    pd.DataFrame(
        [[12, 10, 10]],
        columns=[
            'experience',
            'test_score(out of 10)',
            'interview_score(out of 10)'
        ]
    )
)

print("\nPredicted salary for 2 yrs exp, 9 test score, 6 interview score:")
print(prediction1[0])

print("\nPredicted salary for 12 yrs exp, 10 test score, 10 interview score:")
print(prediction2[0])
```
