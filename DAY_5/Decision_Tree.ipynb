```python
import pandas as pd
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# Load dataset
df = pd.read_csv("/content/titanic.csv")

# Display first 5 rows
print(df.head())

# Select input features
inputs = df[['Pclass', 'Sex', 'Age', 'Fare']]

# Select target variable
target = df['Survived']

# Create Label Encoder object
le_sex = LabelEncoder()

# Convert Sex column into numbers
inputs['Sex_n'] = le_sex.fit_transform(inputs['Sex'])

# Drop original Sex column
inputs = inputs.drop('Sex', axis='columns')

# Fill missing Age values with mean age
inputs['Age'] = inputs['Age'].fillna(inputs['Age'].mean())

# Display processed inputs
print("\nProcessed Inputs:")
print(inputs.head())

# Split dataset into training and testing data
X_train, X_test, y_train, y_test = train_test_split(
    inputs,
    target,
    test_size=0.2,
    random_state=42
)

# Create Decision Tree model
model = tree.DecisionTreeClassifier()

# Train model
model.fit(X_train, y_train)

# Predict sample passenger
# Features:
# [Pclass, Age, Fare, Sex_n]
# male = 1, female = 0

prediction = model.predict(
    pd.DataFrame(
        [[3, 22, 7.25, 1]],
        columns=['Pclass', 'Age', 'Fare', 'Sex_n']
    )
)

print("\nPrediction:")
print(prediction[0])

# Model accuracy
accuracy = model.score(X_test, y_test)

print("\nModel Accuracy:")
print(accuracy)
```
