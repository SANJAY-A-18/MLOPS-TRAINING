```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Load dataset
df = pd.read_csv("insurance_data.csv")

# Display dataset
print(df.head())

# Define input and output
X = df[['age']]
y = df['bought_insurance']

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.1,
    random_state=42
)

# Create Logistic Regression model
model = LogisticRegression()

# Train model
model.fit(X_train, y_train)

# Predict whether person buys insurance
prediction1 = model.predict(
    pd.DataFrame([[35]], columns=['age'])
)

prediction2 = model.predict(
    pd.DataFrame([[22]], columns=['age'])
)

# Predict probabilities
probability1 = model.predict_proba(
    pd.DataFrame([[35]], columns=['age'])
)

probability2 = model.predict_proba(
    pd.DataFrame([[22]], columns=['age'])
)

print("\nPrediction for age 35:")
print(prediction1[0])

print("\nProbability for age 35:")
print(probability1)

print("\nPrediction for age 22:")
print(prediction2[0])

print("\nProbability for age 22:")
print(probability2)

# Model accuracy
accuracy = model.score(X_test, y_test)

print("\nModel Accuracy:")
print(accuracy)
```
