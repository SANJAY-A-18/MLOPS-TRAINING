```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("homeprices.csv")

# Display dataset
print(df)

# Create dummy variables for town column
dummies = pd.get_dummies(df.town, dtype=int)

# Display dummy columns
print(dummies)

# Merge original dataframe with dummy dataframe
merged = pd.concat([df, dummies], axis='columns')

# Display merged dataframe
print(merged)

# Drop original town column
final = merged.drop(['town'], axis='columns')

# Drop one dummy column to avoid dummy variable trap
final = final.drop(['west windsor'], axis='columns')

# Display final dataframe
print(final)

# Define input and output
X = final.drop('price', axis='columns')
y = final.price

# Create Linear Regression model
model = LinearRegression()

# Train model
model.fit(X, y)

# Predict house price
# Example:
# area = 2800
# monroe township = 0
# robinsville = 1 (corrected spelling)

prediction = model.predict(
    pd.DataFrame(
        [[2800, 0, 1]],
        columns=
            ['area',
            'monroe township',
            'robinsville' # Corrected spelling to match training data
        ]
    )
)

print("\nPredicted price:")
print(prediction[0])

# Model accuracy
accuracy = model.score(X, y)

print("\nModel Accuracy:")
print(accuracy)
```
